```python
import orbital_simulator
import orbital_simulator.integrators

import numpy as np
```

```python
def f(t: float, x: np.ndarray):
    return np.array([x[1],x[2],-x[0]-x[1]-x[2]])
```

```python
i = orbital_simulator.integrators.runge_kutta(f=f,x0=np.array([1,1,1]),t0=0)
```

```python
x, t = i.integrate(10)
```

```python
import matplotlib.pyplot as plt

plt.plot(t, x[:,2])
```
